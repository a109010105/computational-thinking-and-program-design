```python
print("~~~~~~~~~~~~~~~~~~~~請注意自助點餐機的回答問題方式喔~~~~~~~~~~~~~~~~~~~~")
print("~~~~~~~~~~~~~~~~~~~~~~點餐回答問題錯誤上限為5次喔~~~~~~~~~~~~~~~~~~~~~~~")
Q=input('機器人店員 : 歡迎光臨豪好吃火鍋!請問有來過我們店裡用餐嗎?(有/沒有) ')
print("客人 : "+Q)
if Q=="有":
  print("機器人店員 : 那點餐方式都相同喔，幫我參考一下菜單決定需要什麼餐點喔!")
elif Q=="沒有" :
  QA=input("機器人店員 : 請問需要幫您介紹嗎?(好/不用) ")
  if QA=="不用":
    print("客人 : "+QA)
    print("機器人店員 : 好的 那參考完菜單再幫我做點餐就可以喔!")
  elif QA=="好":
    print("客人 : "+QA)
    print("機器人店員 : 我們點餐均是以一個麻辣鍋底配上一個不辣的鍋底~")
    print("機器人店員 : 總共有四種湯底，以下為您做介紹與點餐~")
    N=0
    while N<5:
      N+=1
      print("===================================================================")
      print("機器人店員 : 麻辣鍋底的部分我們有兩種,蒙古麻辣鍋比較溫和,椒香鍋比較嗆辣")
      QN=input('機器人店員 : 請問需要哪種? ')
      if QN=="蒙古麻辣鍋":
        N=5
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 好的 請稍後餐點")
      elif QN=="椒香鍋":
        N=5
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 好的 請稍後餐點")
      else:
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 麻辣鍋底的選擇中，沒有這個選項喔! 錯誤次數",N)
      break
    M=0
    while M<5:
      M+=1
      print("===================================================================")
      print("機器人店員 : 我們有兩種不辣的鍋底,柴魚鍋比較清淡,沙茶鍋比較濃郁")
      QN=input('機器人店員 : 請問需要哪種? ')
      if QN=="柴魚鍋":
        M=5
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 好的 已完成主餐部分的點餐")
        print("===================================================================")
      elif QN=="沙茶鍋":
        M=5
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 好的 已完成主餐部分的點餐")
        print("===================================================================")
      else:
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 不辣的鍋底的選擇中，沒有這個選項喔! 錯誤次數",M)
      break
    O=0
    while O<5:
      O+=1
      print("===================================================================")
      print("機器人店員 : 我們有兩種飲品可以選擇,有Vodka跟Tequila")
      QN=input('機器人店員 : 請問需要哪種? ')
      if QN=="Vodka":
        O=5
        print("客人 : 我要一杯"+QN)
        print("機器人店員 : 好的 稍後幫您上飲品")
        print("===================================================================")
      elif QN=="Tequila":
        O=5
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 好的 稍後幫您上飲品")
        print("===================================================================")
      else:
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 飲品的選擇中，沒有這個選項喔! 錯誤次數",O)
      break
    P=0
    while P<5:
      P+=1
      print("===================================================================")
      print("機器人店員 : 甜點的部分有抹茶蛋糕、提拉米蘇可以選擇")
      QN=input('機器人店員 : 請問需要哪種? ')
      if QN=="抹茶蛋糕":
        P=5
        print("客人 : 我要一份"+QN)
        print("機器人店員 : 好的 已完成點餐")
        print("===================================================================")
      elif QN=="提拉米蘇":
        P=5
        print("客人 : 我要一份"+QN)
        print("機器人店員 : 好的 已完成點餐")
        print("===================================================================")
      else:
        print("客人 : 我要一個"+QN)
        print("機器人店員 : 甜點的選擇中，沒有這個選項喔! 錯誤次數",P)
      break
    
  else:
    print("!=!=!=!=!=!=!=!=!=!=!=!=!=!=!=!==!=!=!=!=!=!=!=!=!=!")
    print("機器人店員 : 無法辨識的答案 ! 請重新啟動 ! !")
    print("!=!=!=!=!=!=!=!=!=!=!=!=!=!=!=!==!=!=!=!=!=!=!=!=!=!")
else:
  print("!=!=!=!=!=!=!=!=!=!=!=!=!=!=!=!==!=!=!=!=!=!=!=!=!=!")
  print("機器人店員 : 無法辨識的答案 ! 請重新啟動 ! !")
  print("!=!=!=!=!=!=!=!=!=!=!=!=!=!=!=!==!=!=!=!=!=!=!=!=!=!")
```
